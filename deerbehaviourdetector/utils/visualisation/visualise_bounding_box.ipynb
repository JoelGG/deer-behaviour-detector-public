```python
import numpy as np
import csv
from PIL import Image
```

```python
def crop(frame, bbox):
  width, height = frame.size
  bounding_box = (
    int((bbox[0]) * width),  # x
    int((bbox[1]) * height),  # y
    int((bbox[0] + bbox[2]) * width),  # x + w
    int((bbox[1] + bbox[3]) * height),  # y + h
  )
  # print(bounding_box)
  return frame.crop(bounding_box)
```

```python
frame = Image.open("/Volumes/GHD/dataset/T2801149/frames/1.jpg")
bbox_1 = [0.4369, 0.3574, 0.1463, 0.3824]
bbox_2 = [0.7932, 0.4194, 0.2067, 0.55]
```

```python
crop(frame, bbox_1)
```

```python
crop(frame, bbox_2)
```
