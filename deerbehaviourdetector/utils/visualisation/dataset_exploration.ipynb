```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# mpl.rc('image', cmap='gray')
sns.set_theme()
```

```python
def class_frequency(df, title=""):
  fig, ax = plt.subplots()
  df['target'].value_counts().div(df['target'].value_counts().sum()).plot(ax=ax, kind='bar', ylabel='Relative frequency', xlabel="Class name", color = "grey", title=title)
  plt.show()
```
